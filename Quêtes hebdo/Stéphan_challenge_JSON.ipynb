```python
#on récupère les données pour créer un fichier de data
import json
import requests
import pandas as pd

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/monthlySalesbyCategoryMultiple.json"

# This function gets the file content from the URL
r = requests.get(link)

# You can access to the text content with this function
r.text

data = json.loads( r.text)
```

```python
#affichage des données qui restent un dictionnaire de pluieurs éléments
data
```

```python
#on parcours le dictionnaire à partir d'une base puis on affiche les colonnes souhaitées (Sous forme de chemin via les listes et les dictionnaires)
df_ = pd.json_normalize(data,
                        record_path =['contents','monthlySales'],
                        meta =[['','category'],['','region']]
                        )
```
