```python
import json
import requests
import pandas as pd
from pprint import pprint

# J'ai testé deux versions, une pour récupérer le data entier (link) et l'autre (link2) en calculant le résultat demandé

link = "https://data.nantesmetropole.fr/api/records/1.0/search/?dataset=244400404_stations-velos-libre-service-nantes-metropole-disponibilites&q=&facet=banking&facet=bonus&facet=status&facet=contract_name"


# lien2 qui est un lien avec selection des colonnes et le calcul directement effectué vis à vis de la question principal de la quête

link2 = "https://data.nantesmetropole.fr/api/explore/v2.1/catalog/datasets/244400404_stations-velos-libre-service-nantes-metropole-disponibilites/records?select=address%2Cavailable_bikes%2Fbike_stands%20as%20taux_remplissage"

link3 = "https://data.nantesmetropole.fr/api/explore/v2.1/catalog/datasets/244400404_stations-velos-libre-service-nantes-metropole-disponibilites/exports/json?lang=fr&timezone=Europe%2FBerlin"


#test des 128 lignes en deux fichiers

link4 = "https://data.nantesmetropole.fr/api/explore/v2.1/catalog/datasets/244400404_stations-velos-libre-service-nantes-metropole-disponibilites/records?limit=100&offset=1"
link5 = "https://data.nantesmetropole.fr/api/explore/v2.1/catalog/datasets/244400404_stations-velos-libre-service-nantes-metropole-disponibilites/records?limit=28&offset=2"

# This function gets the file content from the URL
r = requests.get(link)
s = requests.get(link2)
t = requests.get(link3)
u = requests.get(link4)
v = requests.get(link5)

# You can access to the text content with this function
r.text

data = r.json()
data2 = s.json()
data3 = t.json()
data4 = u.json()
data5 = v.json()
```

```python
data4
```

```python
# 
df_velo4 = pd.json_normalize(data4, record_path='results')
df_velo5 = pd.json_normalize(data5, record_path='results')
df_complet = pd.concat([df_velo4,df_velo5],join='inner').reset_index(drop=True)
```

```python
## version 3, création d'une colonne avec le calcul
df_complet['taux_remplissage'] = df_complet['available_bikes']/df_complet['bike_stands']*100
```

```python
# version 3,affichage du résultat demandé 
df_complet[['address','taux_remplissage']].sort_values(by = 'taux_remplissage', ascending=False)
```
