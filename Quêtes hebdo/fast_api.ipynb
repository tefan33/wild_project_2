```python
data["nom_commune_complet"].groupby(data["nom_region"]).agg(["count"])
```

```python
# transformation du calcul de mon nombre de villes par région en dictionnaire
dico_region = data["nom_commune_complet"].groupby(data["nom_region"]).agg(["count"]).to_dict()['count']
```

```python
# on regroupe les départements par régions et par nombre de communes
data[data["nom_region"] == "Auvergne-Rhône-Alpes"].groupby(["nom_departement"]).size().reset_index(name='counts').set_index('nom_departement').to_dict()['counts']
```

```python
# Retourne les communes d'un département (nom_departement) avec leur code commune INSEE
data[data["nom_departement"] == "Ain"][['nom_commune_complet','code_commune_INSEE']].set_index('nom_commune_complet').to_dict()['code_commune_INSEE']
```

```python
# Retourne les informations d'un code Insee (code_commune_INSEE) (code_commune_INSEE, code_postal, latitude, longitude, nom_commune_complet, nom_département et nom_region )
data[data["code_commune_INSEE"] == "74313"].to_dict('records')[0]
```

```python
# transformation du calcul des communes d'un département (nom_departement) avec leur code commune INSEE en liste
liste_code_insee = data[['nom_commune_complet','code_commune_INSEE']].to_dict('split')['data']

```

```python
# on tranforme cette liste de liste en dictionnaire, avec la création d'un dictionnaire
dict_code_insee = {}
for keys, values in liste_code_insee:
    dict_code_insee.update({keys : values})
```
